```python
import openpyxl

# Check if there is an existing data file
existing_data_file = input("Enter the filename of existing data (or leave blank if none): ")
if existing_data_file:
    try:
        existing_wb = openpyxl.load_workbook(existing_data_file)
        sheet = existing_wb.active
        row = sheet.max_row + 1
        print(f"Loaded existing data from {existing_data_file}.")
    except FileNotFoundError:
        print(f"Existing data file {existing_data_file} not found. Starting with new data.")
else:
    # Create a new workbook and select the active sheet
    existing_wb = openpyxl.Workbook()
    sheet = existing_wb.active
    row = 2

# Write column headers if starting with new data
if row == 2:
    sheet["A1"] = "Name"
    sheet["B1"] = "Amount"
    sheet["C1"] = "Date"
    sheet["D1"] = "Time"

# Initialize a dictionary to store spender-wise totals
totals = {}

# Prompt for input until the user enters "done"
while True:
    spender_name = input("Enter the name of the spender (or 'done' to finish): ")
    if spender_name.lower() == "done":
        break

    amount = float(input("Enter the amount of the cost: "))
    date = input("Enter the date of the cost (format: YYYY-MM-DD): ")
    time = input("Enter the time of the cost (format: HH:MM): ")

    # Write data to Excel
    sheet[f"A{row}"] = spender_name
    sheet[f"B{row}"] = amount
    sheet[f"C{row}"] = date
    sheet[f"D{row}"] = time

    # Update spender-wise totals
    if spender_name in totals:
        totals[spender_name] += amount
    else:
        totals[spender_name] = amount

    row += 1

# Sort the data by date
sheet.auto_filter.ref = sheet.dimensions
sheet.auto_filter.add_filter_column(2, [""])
sheet.auto_filter.add_sort_condition("C2:C100")

# Write spender-wise totals
sheet[f"A{row}"] = "Total Amount"
sheet[f"B{row}"] = " "
for spender, total in totals.items():
    row += 1
    sheet[f"A{row}"] = spender
    sheet[f"B{row}"] = total

# Save the workbook
filename = input("Enter the filename to save the data (e.g., expenses.xlsx): ")
existing_wb.save(filename)
print("Data saved successfully!")
```
